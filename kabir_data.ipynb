```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv")
df.head()
```

```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'], format='%Y-%m-%d %H:%M:%S')

df['Date'] = df['Date_Time'].dt.date
df['Time'] = df['Date_Time'].dt.time

df.head()
```

```python
df = df.drop(columns = ["Date_Time"])
```

```python
df.head()
```

```python
grouped = df.groupby('Location')
for location, group in grouped:
    print(f"Location: {location}")
    print(group)
```

```python
dfs_by_location = {location: group for location, group in df.groupby('Location')}
df_new_york = dfs_by_location["New York"]
df_san_diego = dfs_by_location["San Diego"]
df_phoenix = dfs_by_location["Phoenix"]
df_houston = dfs_by_location["Houston"]
df_dallas = dfs_by_location["Dallas"]
df_chicago = dfs_by_location["Chicago"]
df_los_angeles= dfs_by_location["Los Angeles"]
df_philadelphia = dfs_by_location["Philadelphia"]
df_san_antonio = dfs_by_location["San Antonio"]
df_san_jose = dfs_by_location["San Jose"]


df_new_york.to_csv('weather_new_york.csv', index=False)
df_san_diego.to_csv('weather_san_diego.csv', index=False)  
df_phoenix.to_csv('weather_phoenix.csv', index=False)  
df_houston.to_csv('weather_houston.csv', index=False)  
df_dallas.to_csv('weather_dallas.csv', index=False)  
df_chicago.to_csv('weather_chicago.csv', index=False)
df_los_angeles.to_csv('weather_los_angeles.csv', index=False)  
df_philadelphia.to_csv('weather_philadelphia.csv', index=False)  
df_san_antonio.to_csv('weather_san_antonio.csv', index=False)  
df_san_jose.to_csv('weather_san_jose.csv', index=False)  

```
